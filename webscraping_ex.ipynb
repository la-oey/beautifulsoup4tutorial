# Getting Started + Importing Packages

```python
# to install bs4, use the following command in your terminal:
# pip install bs4
from bs4 import BeautifulSoup
import urllib.request
import re
```

# Uploading File to Be Scraped

```python
website = urllib.request.urlopen("https://psychology.ucsd.edu/people/graduate-students.html")
soup = BeautifulSoup(website, 'html.parser')
```

```python
html = open('GraduateStudents.htm', 'rb')
soup2 = BeautifulSoup(html, 'html.parser')
```

# Feeling out the HTML

```python
print(soup.prettify())
```

```python
print(soup.get_text())
```

# Navigating Tags

```python
for h in soup.find_all('h3'):
    print(h.get_text())
```

# Navigating Links

```python
soup.find_all('a')
```

```python
for a in soup.find_all('a'):
    link = a.get('href')
    print(link)
```

```python
for a in soup.find_all('a'):
    link = a.get('href')
    if link is not None and re.match('mailto:(.*?)$', link):
        print(re.match('mailto:(.*?)$', link).group(1))
```

```python
for a in soup.find_all('a'):
    link = a.get('href')
    if link is not None and re.match('^(.*?)\.html$', link):
        print(re.match('^(.*?)\.html$', link).group(0))
```
