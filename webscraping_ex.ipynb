# Now Let's Try Scraping the Pokedex

```python
import csv

user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'
url = "https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
headers = {'User-Agent':user_agent,} 
request = urllib.request.Request(url,None,headers)
website_poke = urllib.request.urlopen(request)
pokedex = BeautifulSoup(website_poke, 'html.parser')
all_pokemon = dict()

with open('my_pokemon.csv', 'w') as csv_file:
    fieldnames = ['Index', 'Pokemon', 'Type']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    
    for table in pokedex.find_all('table'):
        for link in table.find_all('a'):
            linkref = link.get('href')
            if linkref is not None and re.match('^/wiki/(.*?)_\(Pok%C3%A9mon\)$', linkref):
                if linkref not in all_pokemon:
                    all_pokemon[linkref] = linkref
                    name = link["title"]
                    print(name)
                    type1 = link.find_parent('td').findNext('td').findNext('td').find('a')["title"]
                    print(type1)
                    index = link.find_parent('td').findPrevious('td').get_text()[1:5]
                    print(index)
                    writer.writerow({'Index': index, 'Pokemon': name, 'Type': type1})
                    
    
    
    csv_file.close()
```
